```python
import numpy as np
import pandas as pd
```

```python
from numpy import  nan as NA
data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA], [NA, NA, NA], [NA, 6.5, 3.]])
data.index = ['A', 'B', 'C', 'D']
data
```

```python
data.isna()
```

```python
data.dropna()
```

<p>همان‌طور که دیدید، دستور dropna می‌تواند تمام ردیف‌هایی که حتی یک مقدار گم‌شده دارند را حذف کند.</p>

<p>
نکته‌ی مهم

داده‌های گم‌شده در پانداز می‌توانند به شکل‌های مختلفی وجود داشته باشند و همیشه واضح نیستند. برای مثال، ممکن است به عنوان چند اسپیس خالی و یا کاراکتر \xa0 ذخیره شوند. بنابراین سعی کنید همیشه قبل از شروع کار پاکسازی داده‌ها، نگاهی اجمالی به مقادیر ذخیره شده در آن بیندازید.
</p>

```python
data = pd.DataFrame([[1., 6.5, 3., 8., 2.3], [1., NA, NA, 5, 1.4],
                     [NA, NA, 2,  NA, 4.5], [NA, 6.5, 3.,NA, 3.1],
                     [2.5, 8.1, NA, NA, 2], [NA , NA, NA , 4.2, NA],
                     [2, 5.2, 1.8, NA, 2.5]])
data.index = ['A','B','C','D','E','F','G']
data
```

<p>یکی از حالت‌هایی که حذف مقادیر گم‌شده ایده‌ی مناسبی‌ست چنین، شرایطی است. ضمناً می‌توان تنها ردیف‌هایی را نگه داشت که مثلاً حداقل ۳ داده‌ی غیر خالی دارند. برای این کار، به شکل زیر عمل می‌کنیم:</p>

```python
data.dropna(thresh=3)
```

```python
data = pd.DataFrame([[1., 6.5, 3., NA, 2.3], [1., NA, NA, 5, 1.4],
                     [NA, NA, 2,  NA, 4.5], [NA, 6.5, 3.,NA, 3.1],
                     [2.5, 8.1, NA, NA, 2], [NA , NA, 1, NA , NA],
                     [2, 5.2, 1.8, NA, 2.5],  [3, 2.1, 8.1, NA, 2.5],
                     [NA, 8.2, NA, NA, 1.9], [2, NA, 1.5, NA, 7.2]])
data.index = ['A','B','C','D','E','F','G','H','I','J']
data
```

<p>به طور کلی می‌توان گفت اگر ستونی بیش از ۹۰ درصد اطلاعات آن گم‌شده باشد بهتر است آن ستون را به طور کامل از مجموعه‌داده حذف کرد و برای تحلیل‌های آینده از آن استفاده نکرد.</p>

```python
data.isna().sum()
```

```python
data.drop(data.columns[ data.isna().sum()/len(data) >= 0.9] , axis = 1)
```
